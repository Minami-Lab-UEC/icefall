```python
import kenlm
```

```python
models = [kenlm.Model(f"../approx_tgtlens/data/{i}gram.arpa") for i in range(2,6)]
how_much_entropy_per_words = [
    0.41650609460494159,
    0.55819859528638738,
    0.66557475148552303,
    0.72205136160654204,
]
```

```python
sentence = "検 索 対 象 文 書 の 中 の 単 語 の 共 起 関 係 を 元 に そ の 語 単 語 間 の 連 想 関 係 を"
```

```python
outputs = []
for model, how_much_entropy_per_word in zip(models, how_much_entropy_per_words):
    tokens = [
        (*s, t) for s, t in zip(list(model.full_scores(sentence, bos=True, eos=True)), sentence.split(" ") + ["<eos>"])
    ]
    print(tokens)
    how_much_entropy_per_word = 1/how_much_entropy_per_word
    cum_logprob = 0
    out_text = ["<sos>"]
    for ts in tokens:
        logprob, _, _, t = ts
        cum_logprob -= logprob
        while cum_logprob >= how_much_entropy_per_word:
            out_text.append("∅")
            cum_logprob -= how_much_entropy_per_word
        out_text.append(t)
    outputs.append(' '.join(out_text))
```
